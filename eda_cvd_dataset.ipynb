# Exploratory Data Analysis - CVD Dataset

## Import Dataset

```python
import pandas as pd                 # Reading dataset
import numpy as np                  # Numerical computing
import seaborn as sns               # Data visualization
import matplotlib.pyplot as plt     # Data visualization

%matplotlib inline
sns.set(color_codes=True)           # Visual consistency
```

## Input and Inspect the Data

```python
# Read the dataset
data = pd.read_csv('/content/CVD Dataset.csv')
```

```python
# Return number of rows and columns
data.shape
```

```python
# Return first 5 rows of the dataset
data.head()
```

```python
# Return the last 5 rows of a dataset
data.tail()
```

```python
# Return info on dataset
data.info()
```

## Dropping Irrelevant Columns

```python
# Handling irrelevant columns
data = data.drop('Blood Pressure (mmHg)',axis=1)
```

```python
data.info()
```

## Handle Missing Data

```python
# Check for missing value
data.isna().sum()
```

```python
# Check the statistical summary
data.describe()
```

```python
# Separate the categorical column names
categoricals = ['Sex','Smoking Status','Diabetes Status','Physical Activity Level','Family History of CVD','CVD Risk Level','Blood Pressure Category']
```

```python
# Check the statistical summary for categorical data.
data[categoricals].describe()
```

```python
# Handling missing value
for column in data.columns:
    if data[column].dtype == 'object':
        data[column].fillna(data[column].mode()[0], inplace=True)
    else:
        data[column].fillna(data[column].mean(), inplace=True)
```

```python
data.isna().sum()
```

## Handling Duplicate Data

```python
# Check for duplicated Data
check_duplicated = data.duplicated().sum()
print(check_duplicated)
```

## Handling Outliers

```python
# Check the outliers with visualizations using a box plot
sns.boxplot(x=data['CVD Risk Score'])
```

```python
sns.boxplot(x=data['Estimated LDL (mg/dL)'])
```

```python
sns.boxplot(x=data['Waist-to-Height Ratio'])
```

```python
# Check the outliers with IQR Score
Q1 = data[column].quantile(0.25)
Q3 = data[column].quantile(0.75)
IQR = Q3 - Q1
print(IQR)
```

```python
# Handling outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
data = data[(data[column] >= lower_bound) & (data[column] <= upper_bound)]
data.shape
```

## Analyzing the Data

### Univariate Analysis

```python
# Check for categorical data
data["Family History of CVD"].value_counts()
```

```python
# Create a countplot
sns.countplot(data=data,x="Family History of CVD")
plt.title("Number of Clients by Family History of CVD")
plt.ylabel("Number of Clients")
plt.show()
```

```python
# Create a piechart
plt.figure(figsize=[6,6])
blood_pressure_counts = data["Blood Pressure Category"].value_counts(normalize=True)
labels = ["Hypertension Stage 2","Hypertension Stage 1","Elevated","Normal"]
plt.pie(blood_pressure_counts,labels=labels, autopct='%.0f%%')
plt.title("Proportion of Clients by Blood Pressure Category")
plt.show()
```

```python
# Return summary statistics for the numerical data
data['CVD Risk Score'].describe()
```

```python
# Create a histogram
plt.figure(figsize=[8,5])
sns.histplot(data=data,x="CVD Risk Score",bins=20).set(title="Distribution of CVD Risk Score",ylabel="Number of clients")
plt.show()
```

```python
# Create a histogram with kernel density estimation
plt.figure(figsize=[8,5])
sns.histplot(data=data,x="CVD Risk Score",bins=20,kde=True).set(title="Distribution of CVD Risk Score",ylabel="Number of clients")
plt.show()
```

### Bivariate Analysis

```python
# Check the relationship between two numerical data using a scatter plot.
plt.figure(figsize=[8,5])
plt.scatter(data=data,x="Total Cholesterol (mg/dL)",y="CVD Risk Score")
plt.title("Total Cholesterol (mg/dL) vrs CVD Risk Score")
plt.ylabel("CVD Risk Score")
plt.xlabel("Total Cholesterol (mg/dL)")
plt.show()
```

```python
# Check the relationship between two numerical data using a correlation matrix
corr = data[["Total Cholesterol (mg/dL)","CVD Risk Score"]].corr()
corr
```

```python
# Visualize correlation with a heatmap
plt.figure(figsize=[8,5])
sns.heatmap(corr,annot=True)
plt.title("Correlation between Selected Variables")
plt.show()
```

```python
# Check the mean numerical data for each categorical data group
data.groupby('CVD Risk Level')['BMI'].mean()
```

```python
# Plot boxplot
sns.boxplot(data=data,x='CVD Risk Level', y='BMI')
plt.title("Distribution of BMI per CVD Risk Level")
plt.show()
```

```python
# Create histograms to compare distribution
sns.histplot(data,x="BMI",hue="CVD Risk Level",element="step",stat="density")
plt.title("Distribution of BMI per CVD Risk Level")
plt.show()
```

```python
# Check the relationship between two categorical variables
fig, axes = plt.subplots(1,2,figsize=(12,4))
for i,col in enumerate(["Physical Activity Level","Blood Pressure Category"]):
    sns.histplot(data, ax=axes[i],x=col, hue="CVD Risk Level",stat="probability", multiple="fill", shrink=.8,alpha=0.7)
    axes[i].set(title="Claim Probability by "+ col,ylabel=" ",xlabel=" ")
```

### Multivariate Analysis

```python
# Create a new "cvd_rate" column
mapping = {'HIGH':3, 'INTERMEDIARY':2, 'LOW':1}
data['cvd_rate'] = data['CVD Risk Level'].map(mapping)
data['cvd_rate'].value_counts()
```

```python
# Check the relationship between variables in our dataset by using a third variable
bp_cvd = pd.pivot_table(data=data,index='Physical Activity Level',columns='Blood Pressure Category',values='cvd_rate',aggfunc='mean')
bp_cvd
```

```python
# Check the relationship between datasets with a heatmap
plt.figure(figsize=[8,5])
sns.heatmap(bp_cvd,annot=True,cmap='coolwarm',center=0.117)
plt.title("Physical Activity and Blood PRessure")
plt.show()
```
